### 1. Installation
- update system: `sudo apt update`
- install docker: `sudo apt install docker.io`
- start docker: `sudo systemctl start docker`
- enable docker: `sudo systemctl enable docker`

### 2.Checking

- checking version: `docker --version`

### 3. Install NVidia Docker

### 6. Run Docker Image


- `docker run` : run the image
- `-it` : interactive mode
- `-rm` : an option, whcih will destroy the container after you finished and exit it
- `--gpus all` : use all gpus
- `workspace` is folder mounted to container
- `tensorflow/...-gpu` : docker image is going to be used
- `bash` : allow to run the commands




**Current working directories can be interactive under workspace folder in container**
