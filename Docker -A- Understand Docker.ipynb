### Docker Structure



### 1. Image
Docker image contains all data and files for specified environment, like HD image in virtual machine.

image will be `run` in the container

### 2. Container
Container is the place to run docker image, more like virtual machine

### 3. Dockerfile
list all systems files needed, more like a guide on how to setup required environment. 

`docker build` command can build docker image from dockerfile

### 4. Docker Hub
list popular docker images, can be downloaded to local or uploaded from local

### 5. Save TAR file

### 2. Download latest NGINX server image
`docker pull nginx`


### 3. List all images available
`docker images`



### 4. Run docker images with parameters
`docker run -d -p 81:80 nginx`
- -d : run in background
- -p 81:80 : internal port is 81, external port is 80
- --name: myName
- -v : map external folder


### 5. Find running docker containers
`docker ps`



### 6. Test running image

After docker container is started, the port number will show up at the control page, click on it, you will see new webpage from nginx server


webpage from nginx server


### 7. Container ID and Trimmed version

#### 7.1 Docker will show you a long string which is the unique id for this running container


#### 7.2 Docker will only show the first 12 letters when you list all running containers.
